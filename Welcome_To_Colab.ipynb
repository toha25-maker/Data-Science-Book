```
import pandas as pd
import numpy as np

# ধরে নিচ্ছি df_titanic ডেটাফ্রেমটি আগের ধাপে সফলভাবে লোড হয়েছে।
# যদি আপনি নতুন করে Colab সেশন শুরু করেন, তাহলে নিচের লাইনগুলি রান করে ডেটাফ্রেমটি লোড করে নিন:
titanic_data_url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
df_titanic = pd.read_csv(titanic_data_url)

print("--- ডেটাসেটের সংখ্যাসূচক কলামগুলির প্রাথমিক পর্যবেক্ষণ ---")
print(df_titanic.head()) # সংশোধিত লাইন
print("\n")

# 1. প্রাথমিক গাণিতিক ক্রিয়াকলাপ (Basic Arithmetic Operations)
print("--- 1. প্রাথমিক গাণিতিক ক্রিয়াকলাপ ---")

# 'Fare' কলামের গড় ভাড়া
average_fare = df_titanic['Fare'].mean()
print(f"গড় ভাড়া (Average Fare): {average_fare:.2f}")

# 'Age' কলামের সর্বোচ্চ বয়স (NaN মান উপেক্ষা করে)
max_age = df_titanic['Age'].max()
print(f"সর্বোচ্চ বয়স (Maximum Age): {max_age:.0f} (যদি NaN না থাকে)")

# 'Age' কলামের সর্বনিম্ন বয়স (NaN মান উপেক্ষা করে)
min_age = df_titanic['Age'].min()
print(f"সর্বনিম্ন বয়স (Minimum Age): {min_age:.0f} (যদি NaN না থাকে)")

# 2. সংখ্যা তুলনা করা (Number Comparison)
print("\n--- 2. সংখ্যা তুলনা করা ---")

# 18 বছরের বেশি বয়সী যাত্রীদের ফিল্টার করা
# Age কলামে NaN থাকলে, সেই সারিগুলো এই শর্তে False হবে।
adult_passengers = df_titanic[df_titanic['Age'] > 18]
print(f"\n18 বছরের বেশি বয়সী যাত্রীর সংখ্যা: {len(adult_passengers)}")
print("18 বছরের বেশি বয়সী যাত্রীদের প্রথম 5টি সারি (নাম এবং বয়স):")
print(adult_passengers[['Name', 'Age']].head())

# 100 এর বেশি ভাড়া দেওয়া যাত্রীদের ফিল্টার করা
expensive_fare_passengers = df_titanic[df_titanic['Fare'] > 100]
print(f"\n100 এর বেশি ভাড়া দেওয়া যাত্রীর সংখ্যা: {len(expensive_fare_passengers)}")
print("100 এর বেশি ভাড়া দেওয়া যাত্রীদের প্রথম 5টি সারি (নাম এবং ভাড়া):")
print(expensive_fare_passengers[['Name', 'Fare']].head())

# প্রথম শ্রেণীর (Pclass == 1) এবং বেঁচে থাকা (Survived == 1) যাত্রীদের ফিল্টার করা
first_class_survivors = df_titanic[(df_titanic['Pclass'] == 1) & (df_titanic == 1)] # সংশোধিত লাইন
print(f"\nপ্রথম শ্রেণীর বেঁচে থাকা যাত্রীর সংখ্যা: {len(first_class_survivors)}")
print("প্রথম শ্রেণীর বেঁচে থাকা যাত্রীদের প্রথম 5টি সারি:")
print(first_class_survivors.head()) # সংশোধিত লাইন

# 3. সংখ্যা রাউন্ডিং (Number Rounding)
print("\n--- 3. সংখ্যা রাউন্ডিং ---")


# 'Age' কলামের মানগুলিকে নিচের দিকে পূর্ণসংখ্যায় রাউন্ড করা (Floor)
# np.floor() NaN মানগুলিকে NaN হিসেবেই রাখে।
df_titanic['Age_Floor'] = np.floor(df_titanic['Age'])
print("\n'Age' কলামের মানগুলিকে নিচের দিকে পূর্ণসংখ্যায় রাউন্ড করা (np.floor()):")
# শুধুমাত্র NaN নয় এমন মানগুলি দেখানোর জন্য.dropna() ব্যবহার করা হয়েছে
print(df_titanic[['Name', 'Age', 'Age_Floor']].dropna().head())

# 'Fare' কলামের মানগুলিকে উপরের দিকে পূর্ণসংখ্যায় রাউন্ড করা (Ceiling)
df_titanic['Fare_Ceil'] = np.ceil(df_titanic['Fare'])
print("\n'Fare' কলামের মানগুলিকে উপরের দিকে পূর্ণসংখ্যায় রাউন্ড করা (np.ceil()):")
print(df_titanic[['Name', 'Fare', 'Fare_Ceil']].head())

# ডেটাফ্রেমের শেষ অবস্থা (নতুন কলাম সহ)
print("\n--- ডেটাফ্রেমের শেষ অবস্থা (নতুন কলাম সহ) ---")
print(df_titanic.head())
```
